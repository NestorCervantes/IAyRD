```python
# Paso 1. Importar librerías

# Librerías a utilizar
import numpy as np     # generar arreglos
import matplotlib.pyplot as plt     # generar gráficos
from sklearn.model_selection import train_test_split     # poder dividir los conjuntos de datos en proporción
from sklearn.linear_model import LinearRegression     # contiene diferentes modelos de ML
from sklearn.metrics import mean_squared_error     # función para calcular el error cuadrático
```

```python
# Paso 2. Ingreso de datos

np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
# Paso 3. Preprocesamiento de datos

X_train, X_test, y_train, y_test = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=0)
```

```python
# Paso 4. Elección del modelo y entrenamiento del mismo

modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
# Paso 5. Prueba del modelo

# Realizar predicciones con los datos de prueba
predicciones_test = modelo.predict(X_test)

# Coeficientes del modelo
print("Coeficiente de la pendiente:", modelo.coef_[0][0])
print("Intercepto:", modelo.intercept_[0])

# Evaluar el modelo utilizando el error cuadrático medio
mse_test = mean_squared_error(y_test, predicciones_test)
print("Error cuadrático medio (MSE) en conjunto de prueba:", mse_test)

# Visualizar el modelo junto con los datos de prueba
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, predicciones_test, color='red')
plt.title('Regresión Lineal - Conjunto de Prueba')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.show()
```
